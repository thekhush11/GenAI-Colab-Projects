```python
import random

def indirect_shuffle(my_list):
    n = len(my_list)
    print(f"Original list (indirect): {my_list}")

    for i in range(n - 1, 0, -1):
        j = random.randint(0, i)

        my_list[i], my_list[j] = my_list[j], my_list[i]

    print(f"Shuffled list (indirect): {my_list}")
    return my_list
```

outlier using interquartile range(IQR) method

```python
#Minimum, maximum, median, Q1,Q2,IQR
import numpy as np
lst_marks=[45,32,56,75,89,65,65,78,94,34,45,67,56,77,68,98,76,75,45,55,88,87]
minimum,Q1,median,Q3,maximum = np.quantile(lst_marks, [0,0.25,0.5,0.75,1])
print(minimum,Q1,median,Q3,maximum)
```

```python
import pandas as pd
import numpy as np
#Example dataset
data = {'score': [45,48,49,52,56,1000]} #1000 is an outlier
df = pd.DataFrame(data)

#Calculate Z-Score
df['z_score'] = (df['score'] - df ['score'].mean()) / df['score'].std()

#Filter out outliers (|Z| > 3)
df_no_outliers = df[np.abs(df['z_score']) < 3]
print(df_no_outliers)
```
