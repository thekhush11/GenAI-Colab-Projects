```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load Dataset -
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train)
print("*" * 50)
print(y_train)
print("*" * 50)
print(np.unique(y_train))
print("*" * 50)
print(np.unique(x_train))
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Normalize dataset -
x_train = x_train/255.0
x_test = x_test/255.0
print(x_test)
print(np.unique(x_test))
print(x_train)
print(np.unique(x_train))
```

```python
#Visualizing data
i = 90
plt.imshow(x_train[i], cmap="gray")
plt.show()
```

```python
#model = Sequential()
#model.add(Flatten(input_shape=(28,28)))
#model.add(Dense(128, activation="relu"))
#model.add(Dropout(0.2))
#model.add(Dense(10, activation="softmax"))

#Or you can use this format -

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation="relu"),
    Dense(64, activation="relu"),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(10, activation="softmax")
])
```

```python
#Compile the model-
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
#Train a Model -
model.fit(x_train, y_train, epochs=10)
```

```python
pre = model.predict(x_test)
print(pre)
```

```python
pre_label = np.argmax(pre)
print(pre_label)
```

```python
print(np.unique(pre_label))
```

```python
plt.figure(figsize=(10,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_test[i], cmap="gray")

    pre_label = np.argmax(pre[i])
    actual = y_test[i]
    color = "yellow" if pre_label == actual else "red"
    plt.title(f"pre: {pre}\n actual: {actual}", color=color)
    plt.axis("off")
plt.show()
```
