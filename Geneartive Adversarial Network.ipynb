#Generative Adversarial Network -
GANs are a type of Deep Learning model that is used for generate new, senthetic data that is simillar to the real data (data like image, music, text, video)

This model has two Neural Network -
1. Generator
2. Discriminator

> Generator - Generator is basically used to create fake data, that is so realistic that it can fool the discriminator into thinking it is real.

Workflow of Generator -
1. Input
2. Processing (Pass the input data through a series of layers [Dense, Cond transpose layer, Batch norms])
3. Output


> Discriminator - It is to tell us the difference between the real or fake.

Working of GANs-
1. Input Random Noice into the generator.
2. Generator produce fake data.
3. Fake image and real image are passed to the discriminator.
4. Discriminator try to guess which one is the real.
5. Generator update to improve how real the fake looks.
6. Discriminator to update better detect fake.
7. The competition continues until the fake image becomes realistic enough that the discriminator, can not tell the difference.
