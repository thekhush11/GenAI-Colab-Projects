```python
def string_length(s):
    print("string_length called")
    return len(s)
def int_length(i):
    print("int_legth called")
    return string_length(str(i))

def get_length_of_elements(t):
  a=[]
  for x in t:
   if isinstance(x, int): a.append(int_length)
   if isinstance(x, str): a.append(string_length)
  return  length_of_element_iterator_factory(t,a)

class length_of_element_iterator_factory:
    def __init__(self, t, a):
       self.t=t
       self.a=a
    def __iter__(self):
       return length_of_element_iterator(self)

class length_of_element_iterator:
    def __init__(self, obj):
        self.t= obj.t
        self.a= obj.a
        self.index= 0
    def __next__(self):
        if self.index >= len(self.t):
           raise StopIteration
        k= self.a[self.index](self.t[self.index])
        self.index +=1
        return k

x=(10,20,"Good", "Example of functional programming", 30)
b= get_length_of_elements(x)
print()
print()
print("called length of elements")
print()
print()
for i in b: print(i)
```

It is also possible to set a logarithmic scale for one or both axes. This functionality is isfact only one application of a more general transfomation system in Matplotlib. Each of the axes' scales are set seperately using set_xscale and set_yscale methods which accept one parameter (with the value "lpg" in this case):

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic Scale (y)")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x, x**2, x, x**3, lw=2)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18)
ax.set_title("Custom Ticks")
```
